```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
file_path_2020 = '/Users/dmislam/Downloads/Starter_Code 2/Resources/athletic_sales_2020.csv'
df_1 = pd.read_csv(file_path_2020)



# Read the CSV files into DataFrames.
file_path_2021 = '/Users/dmislam/Downloads/Starter_Code 2/Resources/athletic_sales_2021.csv'
df_2 = pd.read_csv(file_path_2021)
```

#### Check the data types of each DataFrame

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
df_combined = pd.concat([df_1, df_2], ignore_index=True)
df_combined.head()
```

```python
# Check if any values are null.
print(df_combined.isnull().sum())
```

```python
# Check the data type of each column

df_combined.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
df_combined['invoice_date'] = pd.to_datetime(df_combined['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
df_combined.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
region_product_sales_groupby = df_combined.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index().rename(columns={'units_sold': 'Total_Products_Sold'})

# Show the top 5 results.
region_product_sales_groupby_top5 = region_product_sales_groupby.sort_values(by='Total_Products_Sold', ascending=False).head()


# Show the top 5 results.
region_product_sales_groupby_top5.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.


# Rename the "units_sold" column to "Total_Products_Sold"
region_product_sales_pivot = df_combined.pivot_table(values='units_sold', index=['region', 'state', 'city'], aggfunc='sum').reset_index().rename(columns={'units_sold': 'Total_Products_Sold'})

# Show the top 5 results.
region_product_sales_pivot_top5 = region_product_sales_pivot.sort_values(by='Total_Products_Sold', ascending=False).head()


region_product_sales_pivot_top5.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
region_total_sales_groupby = df_combined.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index().rename(columns={'total_sales': 'Total Sales'})

# Show the top 5 results.
region_total_sales_groupby_top5 = region_total_sales_groupby.sort_values(by='Total Sales', ascending=False).head()


region_total_sales_groupby_top5.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.


# Optional: Rename the "total_sales" column to "Total Sales"
region_total_sales_pivot = df_combined.pivot_table(values='total_sales', index=['region', 'state', 'city'], aggfunc='sum').reset_index().rename(columns={'total_sales': 'Total Sales'})

# Show the top 5 results.
region_total_sales_pivot_top5 = region_total_sales_pivot.sort_values(by='Total Sales', ascending=False).head()


region_total_sales_pivot_top5.head()
```

### 4. Determine which Retailer had the Most Sales

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"

retailer_total_sales_groupby = df_combined.groupby(['retailer', 'region', 'state', 'city'])['total_sales'].sum().reset_index().rename(columns={'total_sales': 'Total Sales'})

# Show the top 5 results.
retailer_total_sales_groupby_top5 = retailer_total_sales_groupby.sort_values(by='Total Sales', ascending=False).head()

retailer_total_sales_groupby_top5.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.

women_footwear_sales_df = df_combined[df_combined['product'].str.contains("Women's Athletic Footwear")]
women_footwear_sales_df.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"# Using groupby to show the total number of women's athletic footwear sold for each retailer, region, state, and city, and renaming the "units_sold" column to "Womens_Footwear_Units_Sold".
women_footwear_sales_groupby = women_footwear_sales_df.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum().reset_index().rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

# Show the top 5 results.
women_footwear_sales_groupby_top5 = women_footwear_sales_groupby.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head()


women_footwear_sales_groupby_top5.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.


# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
women_footwear_sales_pivot = women_footwear_sales_df.pivot_table(values='units_sold', index=['retailer', 'region', 'state', 'city'], aggfunc='sum').reset_index().rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

# Show the top 5 results.
women_footwear_sales_pivot_top5 = women_footwear_sales_pivot.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head()
women_footwear_sales_pivot_top5.head()
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.


# Optional: Rename the "total_sales" column to "Total Sales"
daily_sales_pivot = women_footwear_sales_df.pivot_table(values='total_sales', index='invoice_date', aggfunc='sum').rename(columns={'total_sales': 'Total Sales'})


# Show the table.

daily_sales_pivot.head(10)


```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
# Sort the resampled pivot table in descending order on "Total Sales".
daily_sales_resampled = daily_sales_pivot.resample('D').sum().sort_values(by='Total Sales', ascending=False)
daily_sales_resampled.head(10)
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.

weekly_sales_resampled = daily_sales_pivot.resample('W').sum().sort_values(by='Total Sales', ascending=False)


# Sort the resampled pivot table in descending order on "Total Sales".
weekly_sales_resampled.head(10)
```
